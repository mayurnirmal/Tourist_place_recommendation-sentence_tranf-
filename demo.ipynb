```python
import os
os.environ["TRANSFORMERS_NO_TF"] = "1"
os.environ["USE_TF"] = "0"
os.environ["TF_CPP_MIN_LOG_LEVEL"] = "3"


import pandas as pd
import numpy as np
import re
from sentence_transformers import SentenceTransformer, util
```

```python
# Load your dataset
df = pd.read_excel("cleaned_file.xlsx")

df.head()

```

```python
# Load your dataset
df = pd.read_csv("cleaned_file.csv")

df.head()
```

```python
#preprocessing and combine text fields
# Fill NaN values with ''
df = df.fillna("")

# Combine descriptive fields to create a single text column
df["combined_text"] = (
    df["Place Name"] + " " +
    df["Description"] + " " +
    df["Predicted_Tag"] + " " +
    df["Best time"] + " " +
    df["city"] + " " +
    df["state"]
)
```

```python
df["combined_text"].head()
```

```python
# Lightweight but powerful embedding model
model = SentenceTransformer("all-MiniLM-L6-v2")
```

```python
place_embeddings = model.encode(df["combined_text"].tolist(), convert_to_tensor=True)
place_embeddings.shape
```

```python
def extract_budget(query):
    # Find numbers like 7000, 5000, 15000
    nums = re.findall(r'\d+', query)
    
    if not nums:
        return None
    
    # take first number, convert to int
    return int(nums[0])
```

```python
def recommend_places(query, top_k=5):
    
    # 1. Extract budget if mentioned
    budget = extract_budget(query)

    # 2. Encode query
    query_embedding = model.encode(query, convert_to_tensor=True)

    # 3. Compute similarity scores
    scores = util.pytorch_cos_sim(query_embedding, place_embeddings)[0].cpu().numpy()

    df["similarity"] = scores

    # 4. Budget filtering with 25% margin
    if budget:
        lower_limit = budget * 0.75   # -25%
        upper_limit = budget * 1.25   # +25%
        
        df_filtered = df[
            df["Package"].astype(float).between(lower_limit, upper_limit)
        ]
    else:
        df_filtered = df.copy()

    # 5. Sort by similarity
    df_sorted = df_filtered.sort_values(by="similarity", ascending=False)

    # 6. Return ALL columns + similarity score
    return df_sorted.head(top_k)
```
